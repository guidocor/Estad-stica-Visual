```python
from scipy import stats
import seaborn as sns
# Adjusted code to set the maximum x-axis range to 100 for all plots and print p-values

# Reset the population mean to 0 for the null hypothesis
hypothetical_mean = 0
observed_statistic = 10

# Generate a skewed population with an exponential distribution and adjust it to have a mean of 0
np.random.seed(0)  # For reproducibility
population_size = 10000
skewed_population = np.random.exponential(scale=30, size=population_size) - 30

# Define different sample sizes to illustrate the concept of sampling distributions
sample_sizes = [10, 50, 100]

# Set up the plots with the same x-axis scale and a maximum range of 100
x_max = 100
fig, axs = plt.subplots(4, 1, figsize=(6, 12), sharex=True)

# Plot the skewed population distribution
sns.histplot(skewed_population, bins=50, kde=True, color="skyblue", ax=axs[0])
axs[0].axvline(hypothetical_mean, color='green', linestyle='solid', linewidth=2)  # Null hypothesis mean
#axs[0].axvline(observed_statistic, color='red', linestyle='dashed', linewidth=2)  # Observed statistic
axs[0].set_title('Población')
axs[0].set_xlabel('Valor')
axs[0].set_ylabel('Frecuencia')
axs[0].set_xlim(left=min(skewed_population), right=x_max)
axs[0].text(0.66,  0.66, ' Media bajo la hipótesis nula', transform=axs[0].transAxes, color='green',
            horizontalalignment='right')
# Add text for legend


# Plot the sampling distributions of the mean for different sample sizes and calculate p-values
p_values = []
for i, size in enumerate(sample_sizes):
    sample_means = [np.mean(np.random.choice(skewed_population, size, replace=False)) for _ in range(10000)]
    sns.histplot(sample_means, bins=30, kde=True, color="skyblue", ax=axs[i+1])
    axs[i+1].axvline(hypothetical_mean, color='green', linestyle='solid', linewidth=2)  # Null hypothesis mean
    axs[i+1].axvline(observed_statistic, color='red', linestyle='dashed', linewidth=2, ymin=0, ymax=0.70)  # Observed statistic
    axs[i+1].set_title(f'Distribución muestral del estadístico (n={size})')
    axs[i+1].set_xlabel('Media')
    axs[i+1].set_ylabel('Frecuencia')
    axs[i+1].set_xlim(left=min(skewed_population), right=x_max)

    axs[i+1].text(0.81, -0.56, ' Valor del estadístico observado', transform=axs[0].transAxes, color='red',
            horizontalalignment='right')
    # Calculate the p-value for each sample size
    # For a one-sided test, we need to check the proportion of the distribution that is as extreme or more extreme than our observed statistic
    p_value = np.mean(np.array(sample_means) >= observed_statistic)
    p_values.append(p_value)

plt.tight_layout()
plt.show()

# Print the p-values
for size, p_value in zip(sample_sizes, p_values):
    print(f"Sample size {size}: p-value = {p_value:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Define the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)

# Calculate the critical value for one-tailed test at 5% significance level
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha)

# Calculate the p-value area
test_statistic = 2  # example test statistic value
p_value_area = 1 - stats.norm.cdf(test_statistic)

# Plot the normal distribution
fig, ax = plt.subplots()
ax.plot(x, y)

# Fill the region for the p-value
px = np.linspace(test_statistic, mu + 3*sigma, 100)
py = stats.norm.pdf(px, mu, sigma)
ax.fill_between(px, py, color='orange', alpha=0.5)

# Add a line for test statistic
ax.axvline(test_statistic, color='green', linestyle='dashed', linewidth=2, label='Test estadístico')

# Add a line for critical value
ax.axvline(critical_value, color='red', linestyle='dashed', linewidth=2, label='Valor crítico')

# Draw arrow for p-value
ax.annotate('valor p', xy=(test_statistic, 0), xytext=(test_statistic + 1, 0.02),
            arrowprops=dict(facecolor='green', shrink=0.05))

# Draw arrow for level of significance (alpha)
ax.annotate('Nivel de sgnificación\n(alpha)', xy=(critical_value, 0), xytext=(critical_value - 1, 0.04),
            arrowprops=dict(facecolor='red', shrink=0.05))

# Draw arrow for region of rejection
ax.annotate('Area de\nrechazo', xy=(critical_value + 1, 0.02), xytext=(critical_value + 2, 0.08),
            arrowprops=dict(facecolor='brown', shrink=0.05))

# Set labels and title
plt.xlabel('Z')
plt.ylabel('Probabilidad')
plt.title('Test de hipótesis')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate distributions
np.random.seed(5)

# Normal distributions
normal_30 = np.random.normal(loc=0, scale=1, size=30)
normal_500 = np.random.normal(loc=0, scale=1, size=475)
normal_500 = np.concatenate((normal_500, np.random.chisquare(df=2, size=25)))
# Non-normal distributions
non_normal_30 = np.random.chisquare(df=2, size=30)
non_normal_500 = np.random.chisquare(df=2, size=500)

# Plot histograms and Q-Q plots
fig, axes = plt.subplots(4, 2, figsize=(7, 9))

# Helper function to plot histogram and QQ-plot
def plot_distribution(data, ax_hist, ax_qq, title):
    # Histogram with normal curve
    ax_hist.hist(data, bins=10, density=True, alpha=0.3, color='g')
    mu, std = stats.norm.fit(data)
    xmin, xmax = ax_hist.get_xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, mu, std)
    ax_hist.plot(x, p, 'k', linewidth=2)

    # Shapiro-Wilk test
    shapiro_test = stats.shapiro(data)
    ax_hist.set_title(f'Histograma de {title}\nShapiro-Wilk p-value: {shapiro_test.pvalue:.3f}')

    # Q-Q plot
    stats.probplot(data, dist="norm", plot=ax_qq)
    ax_qq.set_title(f'Q-Q de {title}')

# Plot normal distribution with n=30
plot_distribution(normal_30, axes[0, 0], axes[0, 1], 'Normal n=30')

# Plot non-normal distribution with n=30
plot_distribution(non_normal_30, axes[1, 0], axes[1, 1], 'No-normal n=30')

# Plot normal distribution with n=500
plot_distribution(normal_500, axes[2, 0], axes[2, 1], 'Normal n=500')

# Plot non-normal distribution with n=500
plot_distribution(non_normal_500, axes[3, 0], axes[3, 1], 'No-normal n=500')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Configuración de los parámetros
mu = 0
sigma = 1
alpha = 0.05
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Función para formatear el eje X
def format_axes(ax):
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda val, pos: '{:.2f}'.format(val)))
    ax.axvline(-2, color='blue', linestyle='--')
    ax.axvline(2, color='blue', linestyle='--')
    ax.text(-2, 0.02, '-2', fontsize=12, color='blue', horizontalalignment='center')
    ax.text(2, 0.02, '2', fontsize=12, color='blue', horizontalalignment='center')
    ax.set_xlim([mu - 4*sigma, mu + 4*sigma])

# Contraste Bilateral
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x, y, label='Distribución Normal')
ax.fill_between(x, 0, y, where=(x <= norm.ppf(alpha/2)) | (x >= norm.ppf(1 - alpha/2)), color='red', alpha=0.5)
ax.text(norm.ppf(alpha/2), 0.01, 'α/2', fontsize=12, color='red', horizontalalignment='center')
ax.text(norm.ppf(1 - alpha/2), 0.01, 'α/2', fontsize=12, color='red', horizontalalignment='center')
ax.set_title('Contraste Bilateral')
ax.legend()
format_axes(ax)
plt.show()

# Contraste Unilateral Izquierdo
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x, y, label='Distribución Normal')
ax.fill_between(x, 0, y, where=(x <= norm.ppf(alpha)), color='red', alpha=0.5)
ax.text(norm.ppf(alpha), 0.01, 'α', fontsize=12, color='red', horizontalalignment='center')
ax.set_title('Contraste Unilateral Izquierdo')
ax.legend()
format_axes(ax)
plt.show()

# Contraste Unilateral Derecho
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x, y, label='Distribución Normal')
ax.fill_between(x, 0, y, where=(x >= norm.ppf(1 - alpha)), color='red', alpha=0.5)
ax.text(norm.ppf(1 - alpha), 0.01, 'α', fontsize=12, color='red', horizontalalignment='center')
ax.set_title('Contraste Unilateral Derecho')
ax.legend()
format_axes(ax)
plt.show()

# Gráfica para el ejemplo adicional del contraste unilateral izquierdo
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x, y, label='Distribución Normal')
ax.fill_between(x, 0, y, where=(x <= norm.ppf(1 - alpha)), color='red', alpha=0.5)
ax.axvline(2, color='blue', linestyle='--')
ax.text(2, 0.01, '2', fontsize=12, color='blue', horizontalalignment='center')
ax.text(norm.ppf(1 - alpha), 0.02, 'p = 0.975', fontsize=12, color='black', horizontalalignment='center')
ax.set_title('Contraste Unilateral Izquierdo (Ejemplo)')
ax.legend()
format_axes(ax)
plt.show()

# Gráfica para el ejemplo adicional del contraste unilateral derecho
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x, y, label='Distribución Normal')
ax.fill_between(x, 0, y, where=(x >= norm.ppf(1 - alpha)), color='red', alpha=0.5)
ax.axvline(2, color='blue', linestyle='--')
ax.text(2, 0.01, '2', fontsize=12, color='blue', horizontalalignment='center')
ax.text(norm.ppf(1 - alpha), 0.02, 'p = 0.025', fontsize=12, color='black', horizontalalignment='center')
ax.set_title('Contraste Unilateral Derecho (Ejemplo)')
ax.legend()
format_axes(ax)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button

# Set up the initial parameters
np.random.seed(0)
n_samples = 1000
sample_size = 30
population = np.random.exponential(scale=1.0, size=100000)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))
plt.subplots_adjust(bottom=0.25)

# Generate initial sample means
sample_means = [np.mean(np.random.choice(population, size=sample_size)) for _ in range(n_samples)]

# Plot the histogram
n, bins, patches = ax.hist(sample_means, bins=50, density=True, alpha=0.7)
ax.set_xlabel('Sample Mean')
ax.set_ylabel('Frequency')
ax.set_title(f'Distribution of Sample Means (n={sample_size})')

# Add a line for the theoretical normal distribution
x = np.linspace(min(sample_means), max(sample_means), 100)
ax.plot(x, np.exp(-(x - np.mean(sample_means))**2 / (2 * np.var(sample_means)))
        / np.sqrt(2 * np.pi * np.var(sample_means)),
        linewidth=2, color='r')

# Create slider for sample size
ax_slider = plt.axes([0.2, 0.1, 0.6, 0.03])
slider = Slider(ax_slider, 'Sample Size', 2, 100, valinit=sample_size, valstep=1)

# Create reset button
ax_button = plt.axes([0.8, 0.025, 0.1, 0.04])
button = Button(ax_button, 'Reset')

# Update function for slider
def update(val):
    sample_size = int(slider.val)
    sample_means = [np.mean(np.random.choice(population, size=sample_size)) for _ in range(n_samples)]
    ax.clear()
    n, bins, patches = ax.hist(sample_means, bins=50, density=True, alpha=0.7)
    ax.set_xlabel('Sample Mean')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Distribution of Sample Means (n={sample_size})')
    x = np.linspace(min(sample_means), max(sample_means), 100)
    ax.plot(x, np.exp(-(x - np.mean(sample_means))**2 / (2 * np.var(sample_means)))
            / np.sqrt(2 * np.pi * np.var(sample_means)),
            linewidth=2, color='r')
    fig.canvas.draw_idle()

# Reset function for button
def reset(event):
    slider.reset()

# Connect the slider and button to their functions
slider.on_changed(update)
button.on_clicked(reset)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 1. Configuración de la semilla para reproducibilidad
np.random.seed(42)

# 2. Generación de Datos
n = 500  # Número de observaciones

# Edad: Distribución uniforme entre 18 y 70 años
edad = np.random.randint(18, 70, size=n)

# Tipo de Publicidad: 'Digital' o 'Tradicional'
tipo_publicidad = np.random.choice(['Digital', 'Tradicional'], size=n)

# Ingresos: Distribución normal, media=50k, sd=15k
ingresos = np.random.normal(50000, 15000, size=n)
ingresos = np.clip(ingresos, 10000, None)  # Evitar ingresos negativos

# 3. Codificación de Variables Categóricas
# Convertir 'tipo_publicidad' a variable dummy
tipo_publicidad_dummy = np.where(tipo_publicidad == 'Tradicional', 1, 0)

# 4. Definición de la Relación para Intención de Compra
# Ajustamos beta_4 para hacer más notorio el efecto de la interacción
beta_0 = 10
beta_1 = 0.05      # Efecto de la edad
beta_2 = 0.0003    # Efecto de los ingresos
beta_3 = 5         # Diferencia base por tipo de publicidad
beta_4 = 0.25      # Interacción: efecto adicional de la edad para publicidad tradicional
epsilon = np.random.normal(0, 2, size=n)  # Error aleatorio

# Cálculo de la intención de compra
intencion_compra = (beta_0 +
                    beta_1 * edad +
                    beta_2 * ingresos +
                    beta_3 * tipo_publicidad_dummy +
                    beta_4 * tipo_publicidad_dummy * edad +
                    epsilon)

# 5. Creación del DataFrame
datos = pd.DataFrame({
    'IntencionCompra': intencion_compra,
    'Edad': edad,
    'TipoPublicidad': tipo_publicidad,
    'Ingresos': ingresos
})

# 6. Análisis de la Interacción utilizando un Modelo Lineal
modelo = ols('IntencionCompra ~ Edad * TipoPublicidad + Ingresos', data=datos).fit()
print(modelo.summary())

# 7. Visualización de la Interacción
plt.figure(figsize=(12, 8))
sns.scatterplot(data=datos, x='Edad', y='IntencionCompra', hue='TipoPublicidad', alpha=0.5, palette={'Digital':'#1f77b4', 'Tradicional':'#ff7f0e'})

# Crear líneas de regresión para cada tipo de publicidad
edad_range = np.linspace(datos['Edad'].min(), datos['Edad'].max(), 100)

# Para publicidad digital
intencion_digital = (beta_0 + beta_1 * edad_range +
                     beta_2 * ingresos.mean() +
                     beta_3 * 0 +
                     beta_4 * 0 * edad_range)

# Para publicidad tradicional
intencion_tradicional = (beta_0 + beta_1 * edad_range +
                         beta_2 * ingresos.mean() +
                         beta_3 * 1 +
                         beta_4 * 1 * edad_range)

plt.plot(edad_range, intencion_digital, color='#1f77b4', label='Digital', linewidth=2.5)
plt.plot(edad_range, intencion_tradicional, color='#ff7f0e', label='Tradicional', linewidth=2.5)

plt.title('Interacción entre Tipo de Publicidad y Edad en la Intención de Compra', fontsize=16)
plt.xlabel('Edad', fontsize=14)
plt.ylabel('Intención de Compra', fontsize=14)
plt.legend(title='Tipo de Publicidad', fontsize=12, title_fontsize=12)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo para los gráficos
sns.set(style="whitegrid")

# 1. Gráfico de Estudios Entre-Sujetos
def plot_between_subjects():
    np.random.seed(42)  # Para reproducibilidad

    # Generación de datos
    grupo = ['Grupo 1'] * 50 + ['Grupo 2'] * 50
    # Grupo 1: Media 1, alta variabilidad
    datos_grupo1 = np.random.normal(loc=1, scale=0.5, size=50)
    # Grupo 2: Media 2, alta variabilidad
    datos_grupo2 = np.random.normal(loc=2, scale=0.5, size=50)
    valores = np.concatenate([datos_grupo1, datos_grupo2])

    df_between = pd.DataFrame({
        'Grupo': grupo,
        'Valor': valores
    })

    # Cálculo de medias
    medias = df_between.groupby('Grupo')['Valor'].mean().reset_index()

    # Creación del gráfico
    plt.figure(figsize=(10, 6))
    sns.stripplot(x='Grupo', y='Valor', data=df_between, jitter=True, alpha=0.6, palette=['#1f77b4', '#ff7f0e'])

    # Añadir medias
    sns.pointplot(x='Grupo', y='Valor', data=medias, join=False, color='black', markers='D', scale=1.5, ci=None)

    # Añadir línea conectando las medias
    plt.plot([0, 1], medias['Valor'], color='black', linestyle='--', linewidth=2)

    plt.title('Estudios Entre-Sujetos', fontsize=16)
    plt.xlabel('Grupo', fontsize=14)
    plt.ylabel('Valor', fontsize=14)
    plt.ylim(0, 4)
    plt.show()

# 2. Gráfico de Estudios Intra-Sujetos
def plot_within_subjects():
    np.random.seed(42)  # Para reproducibilidad

    # Generación de datos
    n = 30  # Número de sujetos
    sujetos = [f'Sujeto {i+1}' for i in range(n)]

    # Momento 1: Media 1, alta variabilidad
    momento1 = np.random.normal(loc=1, scale=0.8, size=n)
    # Momento 2: Media 2, alta variabilidad
    momento2 = np.random.normal(loc=2, scale=0.8, size=n)

    df_within = pd.DataFrame({
        'Sujeto': sujetos,
        'Momento1': momento1,
        'Momento2': momento2
    })

    # Creación del gráfico
    plt.figure(figsize=(12, 8))

    for idx in range(n):
        plt.plot(['Momento 1', 'Momento 2'], df_within.loc[idx, ['Momento1', 'Momento2']], marker='o', color='gray', alpha=0.5)

    # Promedios
    media1 = df_within['Momento1'].mean()
    media2 = df_within['Momento2'].mean()
    plt.plot(['Momento 1', 'Momento 2'], [media1, media2], marker='D', color='black', linestyle='--', linewidth=2, markersize=8, label='Media')

    plt.title('Estudios Intra-Sujetos', fontsize=16)
    plt.xlabel('Momento', fontsize=14)
    plt.ylabel('Valor', fontsize=14)
    plt.legend()
    plt.ylim(-1, 4)
    plt.show()

# 3. Ejecución de los Gráficos
def main():
    plt.figure(figsize=(14, 6))

    # Gráfico Entre-Sujetos
    plt.subplot(1, 2, 1)
    plot_between_subjects()

    # Gráfico Intra-Sujetos
    plt.subplot(1, 2, 2)
    plot_within_subjects()

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parámetros para la gráfica
np.random.seed(42)  # Reproducibilidad
sample_size = 10000

# Función para generar datos de dos grupos con una diferencia d, dado una desviación típica (std)
def generate_data(d, std=1.0):
    # Grupo A con media = 0
    groupA = np.random.normal(loc=0, scale=std, size=sample_size)
    # Grupo B con media = d * std
    groupB = np.random.normal(loc=d * std, scale=std, size=sample_size)
    return groupA, groupB

# Escenarios a graficar
d_values = [0.1, 0.5, 1.0]
variabilities = [1.0, 0.5]  # Alta vs Baja variabilidad

# Crear subplots: 3 filas (para cada d), 2 columnas (alta y baja variabilidad)
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 18), sharex=True, sharey=True)
axes = axes.flatten()

for i, d_val in enumerate(d_values):
    for j, var in enumerate(variabilities):
        # Índice del subplot
        ax_idx = i * 2 + j
        ax = axes[ax_idx]

        # Generar datos
        groupA, groupB = generate_data(d_val, std=var)

        # Calcular medias
        meanA = np.mean(groupA)
        meanB = np.mean(groupB)
        diff_mean = meanB - meanA

        # Graficar las densidades con mayor transparencia
        sns.kdeplot(groupA, ax=ax, shade=True, color='blue', alpha=0.2, label='Grupo A')
        sns.kdeplot(groupB, ax=ax, shade=True, color='red', alpha=0.2, label='Grupo B')

        # Dibujar líneas verticales para las medias
        ax.axvline(meanA, color='blue', linestyle='--', linewidth=2, label='Media A')
        ax.axvline(meanB, color='red', linestyle='--', linewidth=2, label='Media B')

        # Determinar el límite inferior para la línea horizontal (usaremos el 15% del rango y para mayor altura)
        y_min, y_max = ax.get_ylim()
        y_pos = y_min + 0.15 * (y_max - y_min)

        # Dibujar línea horizontal que conecta las medias, ahora de color negro
        ax.hlines(y=y_pos, xmin=meanA, xmax=meanB, colors='black', linestyles='-', linewidth=2)

        # Añadir una flecha que indique la diferencia de medias
        ax.annotate('', xy=(meanB, y_pos), xytext=(meanA, y_pos),
                    arrowprops=dict(arrowstyle='<->', color='black', lw=2))

        # Anotar el valor de la diferencia de medias
        ax.text((meanA + meanB) / 2, y_pos + 0.02 * (y_max - y_min),
                f'Diferencia: {diff_mean:.2f}', color='black', ha='center', va='bottom', fontsize=10)

        # Configurar títulos y etiquetas
        var_label = 'Alta Variabilidad' if var == 1.0 else 'Baja Variabilidad'
        ax.set_title(f"d = {d_val}, {var_label}", fontsize=14)
        ax.set_xlabel("Valor de la variable", fontsize=12)
        ax.set_ylabel("Densidad", fontsize=12)
        ax.legend(loc='upper right', fontsize=10)

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Configuración de estilo
sns.set(style="whitegrid")

# Función para generar datos simulados con d de Cohen = 0.3 y variabilidad específica
def generar_datos(mean_control, sd_control, d, sd_experimental, n=100):
    """
    Genera datos para grupos control y experimental.

    Parámetros:
    - mean_control: Media del grupo control
    - sd_control: Desviación estándar del grupo control
    - d: Tamaño de efecto de Cohen
    - sd_experimental: Desviación estándar del grupo experimental
    - n: Número de observaciones por grupo

    Retorna:
    - control: Datos del grupo control
    - experimental: Datos del grupo experimental
    """
    # Calcular la media del grupo experimental basada en d de Cohen
    mean_experimental = mean_control + d * sd_control
    control = np.random.normal(mean_control, sd_control, n)
    experimental = np.random.normal(mean_experimental, sd_experimental, n)
    return control, experimental

# Configuración de los estudios
estudios = {
    'Estudio A (0-10)': {
        'mean_control': 5,
        'sd_control': 1.5,
        'd': 0.3,
        'sd_experimental': 2.0
    },
    'Estudio B (0-100)': {
        'mean_control': 50,
        'sd_control': 10,
        'd': 0.3,
        'sd_experimental': 12
    },
    'Estudio C (-5 a 5)': {
        'mean_control': 0,
        'sd_control': 2,
        'd': 0.3,
        'sd_experimental': 1.5
    }
}

# Generar datos
datos = []
np.random.seed(42)  # Para reproducibilidad

for estudio, params in estudios.items():
    control, experimental = generar_datos(
        params['mean_control'],
        params['sd_control'],
        params['d'],
        params['sd_experimental']
    )
    df_control = pd.DataFrame({
        'Puntuación': control,
        'Grupo': 'Control',
        'Estudio': estudio
    })
    df_experimental = pd.DataFrame({
        'Puntuación': experimental,
        'Grupo': 'Experimental',
        'Estudio': estudio
    })
    datos.append(df_control)
    datos.append(df_experimental)

datos_combinados = pd.concat(datos, ignore_index=True)

# Crear el gráfico
g = sns.FacetGrid(datos_combinados, col="Estudio", hue="Grupo", sharex=False, sharey=False, height=5, aspect=1.2)
g.map(sns.kdeplot, "Puntuación", shade=True, alpha=0.5)
g.add_legend()

# Añadir líneas de media y anotaciones
for ax, (estudio, params) in zip(g.axes.flat, estudios.items()):
    control_mean = params['mean_control']
    experimental_mean = control_mean + params['d'] * params['sd_control']
    control_sd = params['sd_control']
    experimental_sd = params['sd_experimental']
    d_cohen = params['d']
    mean_diff = experimental_mean - control_mean

    # Líneas verticales para medias
    ax.axvline(control_mean, color='blue', linestyle='--', label='Media Control')
    ax.axvline(experimental_mean, color='orange', linestyle='--', label='Media Experimental')


    # Anotar diferencia de medias y d de Cohen en la parte inferior
    ax.annotate(f'D de Cohen: {d_cohen}\nDiferencia de Medias: {mean_diff:.2f}\n\nMedia Control: {control_mean:.2f}\nSD Control: {control_sd}\nMedia Experimental: {experimental_mean:.2f}\nSD Experimental: {experimental_sd}',
                xy=(0.5, 0.15), xycoords='axes fraction',
                ha='center', va='center',
                fontsize=10,
                bbox=dict(boxstyle="round,pad=0.3", fc="yellow", alpha=1))

    # Ajustar límites para espacio de anotaciones
    ax.set_ylim(bottom=ax.get_ylim()[0] - ax.get_ylim()[0]*0.1)

# Ajustar títulos y etiquetas
g.set_titles("{col_name}")
g.set_axis_labels("Puntuación", "Densidad")

# Ajustar espacio para anotaciones
plt.subplots_adjust(bottom=0.2)

plt.suptitle('Comparación de Distribuciones con Variabilidad de Grupos Control y Experimental en Diferentes Escalas', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import random

def generar_estudios(num_estudios, efecto_combinado, ci_combinado, variabilidad, seed=None):
    """
    Genera tamaños de efecto y sus intervalos de confianza para estudios individuales.

    Parámetros:
    - num_estudios: Número de estudios individuales.
    - efecto_combinado: Tamaño de efecto combinado final.
    - ci_combinado: Lista con los límites inferior y superior del intervalo de confianza del efecto combinado.
    - variabilidad: 'alta' o 'baja', determina la dispersión de los estudios individuales.
    - seed: Semilla para reproducibilidad.

    Retorna:
    - tamanios_efecto: Lista de tamaños de efecto para cada estudio.
    - ci_inferior: Lista de límites inferiores del intervalo de confianza para cada estudio.
    - ci_superior: Lista de límites superiores del intervalo de confianza para cada estudio.
    """
    if seed is not None:
        np.random.seed(seed)
        random.seed(seed)

    tamanios_efecto = []
    ci_inferior = []
    ci_superior = []

    # Definir parámetros de dispersión según la variabilidad
    if variabilidad == 'alta':
        # Mayor desviación estándar y mayor amplitud de CIs
        efecto_std = 0.3
        ci_width_std = 0.15
    elif variabilidad == 'baja':
        # Menor desviación estándar y menor amplitud de CIs
        efecto_std = 0.1
        ci_width_std = 0.05
    else:
        raise ValueError("Variabilidad debe ser 'alta' o 'baja'.")

    # Generar tamaños de efecto individuales alrededor del efecto combinado
    for _ in range(num_estudios):
        # Tamaño de efecto con desviación estándar
        efecto = np.random.normal(loc=efecto_combinado, scale=efecto_std)
        tamanios_efecto.append(efecto)

        # Amplitud del intervalo de confianza
        ci_half_width = abs(np.random.normal(loc=0.1, scale=ci_width_std))

        # Límites inferior y superior
        ci_inf = efecto - ci_half_width
        ci_sup = efecto + ci_half_width
        ci_inferior.append(ci_inf)
        ci_superior.append(ci_sup)

    return tamanios_efecto, ci_inferior, ci_superior

def grafico_forest(tamanios_efecto, ci_inferior, ci_superior, efecto_combinado, ci_combinado, titulo, ax, study_labels):
    """
    Crea un forest plot.

    Parámetros:
    - tamanios_efecto: Lista de tamaños de efecto para cada estudio.
    - ci_inferior: Lista de límites inferiores del intervalo de confianza para cada estudio.
    - ci_superior: Lista de límites superiores del intervalo de confianza para cada estudio.
    - efecto_combinado: Tamaño de efecto combinado final.
    - ci_combinado: Lista con los límites inferior y superior del intervalo de confianza del efecto combinado.
    - titulo: Título del gráfico.
    - ax: Objeto Axes de Matplotlib donde se dibujará el gráfico.
    - study_labels: Lista de etiquetas para cada estudio.
    """
    n = len(tamanios_efecto)

    # Posiciones en el eje Y (invertido para que el primer estudio esté arriba)
    y_positions = list(range(n, 0, -1))

    # Graficar los intervalos de confianza y los tamaños de efecto individuales usando errorbar
    ax.errorbar(
        tamanios_efecto, y_positions,
        xerr=[ [tamanios_efecto[i] - ci_inferior[i] for i in range(n)],
               [ci_superior[i] - tamanios_efecto[i] for i in range(n)]],
        fmt='o', color='blue', ecolor='black', elinewidth=1.5, capsize=4, label='Estudios individuales'
    )

    # Añadir las etiquetas de los estudios y los valores de efecto
    for i, (x, y) in enumerate(zip(tamanios_efecto, y_positions)):
        ax.text(x, y + 0.1, f"{x:.2f} [{ci_inferior[i]:.2f}, {ci_superior[i]:.2f}]",
                ha='center', va='bottom', fontsize=9)

    # Graficar el intervalo de confianza del efecto combinado
    ax.plot(ci_combinado, [0, 0], color="red", lw=2, label="Intervalo combinado")

    # Dibujar un diamante para el efecto combinado
    diamond = patches.Polygon([
        [ci_combinado[0], 0],
        [efecto_combinado, -0.3],
        [ci_combinado[1], 0],
        [efecto_combinado, 0.3]
    ], closed=True, color='red', alpha=0.6, label='Efecto combinado')
    ax.add_patch(diamond)

    # Añadir la anotación del efecto combinado
    ax.text(efecto_combinado, 0.5, f"{efecto_combinado:.2f} [{ci_combinado[0]:.2f}, {ci_combinado[1]:.2f}]",
            ha='center', va='bottom', color='red', fontsize=10, fontweight='bold')

    # Formato del gráfico
    ax.set_yticks(y_positions)
    ax.set_yticklabels(study_labels)
    ax.set_xlabel("Tamaño del efecto", fontsize=12)
    ax.set_title(titulo, fontsize=14, fontweight='bold')
    ax.axvline(0, color="gray", linestyle="--", lw=1)
    ax.grid(True, axis='x', linestyle='--', alpha=0.7)

    # Ajustar límites del eje y para espacio del efecto combinado
    ax.set_ylim(-1, n + 1)

    # Ajustes estéticos
    ax.tick_params(axis='both', which='major', labelsize=10)
    ax.legend(loc='upper right', fontsize=10)

def main():
    """
    Función principal para generar y plotear forest plots de manera programática.
    """
    # Parámetros de entrada
    num_estudios = 10  # Número de estudios individuales
    variabilidad = 'alta'  # 'alta' o 'baja'
    seed = 42  # Semilla para reproducibilidad (opcional)

    # Definir los efectos combinados para cada caso
    casos = [
        {
            'efecto_combinado': 0.2,
            'ci_combinado': [0.1, 0.3],
            'titulo': 'Meta-análisis: efecto combinado significativo'
        },
        {
            'efecto_combinado': 0.05,
            'ci_combinado': [-0.05, 0.15],
            'titulo': 'Meta-análisis: efecto combinado no significativo'
        }
    ]

    # Crear figura con subplots para cada caso
    fig, axes = plt.subplots(1, len(casos), figsize=(14, 8), sharex=True)

    for idx, caso in enumerate(casos):
        # Generar datos de estudios individuales
        tamanios_efecto, ci_inferior, ci_superior = generar_estudios(
            num_estudios=num_estudios,
            efecto_combinado=caso['efecto_combinado'],
            ci_combinado=caso['ci_combinado'],
            variabilidad=variabilidad,
            seed=seed  # Opcional, para reproducibilidad
        )

        # Crear etiquetas para estudios
        study_labels = [f"Estudio {i+1}" for i in range(num_estudios)]

        # Graficar el forest plot
        grafico_forest(
            tamanios_efecto=tamanios_efecto,
            ci_inferior=ci_inferior,
            ci_superior=ci_superior,
            efecto_combinado=caso['efecto_combinado'],
            ci_combinado=caso['ci_combinado'],
            titulo=caso['titulo'],
            ax=axes[idx],
            study_labels=study_labels
        )

    # Ajustes finales
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns

# Configuración de estilo utilizando Seaborn
sns.set_style("whitegrid")

# Función para simular una diferencia de medias y calcular su IC
def simular_ic(true_diff, n=30, sd=1, alpha=0.05):
    """
    Simula una muestra y calcula el intervalo de confianza para la diferencia de medias.

    Parámetros:
    - true_diff: Diferencia verdadera de medias.
    - n: Tamaño de la muestra por grupo.
    - sd: Desviación estándar de ambos grupos.
    - alpha: Nivel de significancia.

    Retorna:
    - IC_lower: Límite inferior del IC.
    - IC_upper: Límite superior del IC.
    - diff: Diferencia observada de medias.
    """
    # Generar datos para el grupo control y experimental
    control = np.random.normal(loc=0, scale=sd, size=n)
    experimental = np.random.normal(loc=true_diff, scale=sd, size=n)

    # Calcular medias y desviaciones estándar
    mean_ctrl = np.mean(control)
    mean_exp = np.mean(experimental)
    sd_ctrl = np.std(control, ddof=1)
    sd_exp = np.std(experimental, ddof=1)

    # SD combinada (Pooled SD)
    pooled_sd = np.sqrt(((n - 1)*sd_ctrl**2 + (n - 1)*sd_exp**2) / (2*n - 2))

    # Diferencia de medias
    diff = mean_exp - mean_ctrl

    # Error estándar de la diferencia
    se_diff = pooled_sd * np.sqrt(2/n)

    # Grados de libertad
    df = 2*n - 2

    # Valor crítico t para el IC del 95%
    t_crit = stats.t.ppf(1 - alpha/2, df)

    # Intervalo de confianza
    IC_lower = diff - t_crit * se_diff
    IC_upper = diff + t_crit * se_diff

    return IC_lower, IC_upper, diff

# Función para generar múltiples ICs
def generar_ic_multiple(true_diff, num_sim=30, n=30, sd=1, alpha=0.05):
    """
    Genera múltiples intervalos de confianza para un verdadero valor de diferencia de medias.

    Parámetros:
    - true_diff: Diferencia verdadera de medias.
    - num_sim: Número de simulaciones.
    - n: Tamaño de la muestra por grupo.
    - sd: Desviación estándar de ambos grupos.
    - alpha: Nivel de significancia.

    Retorna:
    - Listas de límites inferiores y superiores de los ICs.
    """
    IC_lower_list = []
    IC_upper_list = []
    diffs = []
    for _ in range(num_sim):
        IC_lower, IC_upper, diff = simular_ic(true_diff, n, sd, alpha)
        IC_lower_list.append(IC_lower)
        IC_upper_list.append(IC_upper)
        diffs.append(diff)
    return IC_lower_list, IC_upper_list, diffs

# Configuración de los escenarios
escenarios = {
    'Escenario 1: Parámetro = 0.5': 0.5,
    'Escenario 2: Parámetro = 0': 0
}

# Generar los ICs para cada escenario
resultados = {}
np.random.seed(42)  # Para reproducibilidad

for nombre, valor in escenarios.items():
    IC_lower, IC_upper, diffs = generar_ic_multiple(true_diff=valor, num_sim=30, n=30, sd=1, alpha=0.05)
    resultados[nombre] = {
        'IC_lower': IC_lower,
        'IC_upper': IC_upper,
        'diffs': diffs
    }

# Función para plotear los ICs
def plot_ic(resultados, escenarios):
    """
    Plotea los intervalos de confianza para cada escenario.

    Parámetros:
    - resultados: Diccionario con los ICs y diferencias.
    - escenarios: Diccionario con los nombres y valores verdaderos.
    """
    num_escenarios = len(escenarios)
    fig, axes = plt.subplots(1, num_escenarios, figsize=(20, 10), sharey=False)

    if num_escenarios == 1:
        axes = [axes]

    for ax, (nombre, valor) in zip(axes, escenarios.items()):
        IC_lower = resultados[nombre]['IC_lower']
        IC_upper = resultados[nombre]['IC_upper']
        diffs = resultados[nombre]['diffs']

        # Crear una lista de posiciones
        posiciones = np.arange(1, len(IC_lower)+1)

        # Plotear líneas para cada IC
        for pos, lower, upper in zip(posiciones, IC_lower, IC_upper):
            ax.plot([lower, upper], [pos, pos], color='blue', marker='o', markersize=4)

        # Línea del valor verdadero
        ax.axvline(x=valor, color='red', linestyle='--', label='Valor Verdadero')

        # Configuración del eje
        ax.set_title(nombre, fontsize=16)
        ax.set_xlabel('Diferencia de Medias', fontsize=14)
        ax.set_ylabel('Simulación', fontsize=14)
        ax.set_ylim(0, len(IC_lower)+1)
        ax.set_xlim(min(IC_lower) - 0.5, max(IC_upper) + 0.5)
        ax.invert_yaxis()
        ax.legend(fontsize=12)

        # Agregar etiquetas de texto para cada línea si es necesario
        # Por simplicidad, se omiten para mantener la claridad

    plt.suptitle('El Baile de los Intervalos de Confianza para la Diferencia de Medias (30 Simulaciones)', fontsize=20, y=0.95)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Plotear los ICs
plot_ic(resultados, escenarios)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns

# Configuración de estilo utilizando Seaborn
sns.set_style("whitegrid")

# Función para simular una diferencia de medias y calcular su IC
def simular_ic(true_diff, n=30, sd=1, alpha=0.05):
    """
    Simula una muestra y calcula el intervalo de confianza para la diferencia de medias.

    Parámetros:
    - true_diff: Diferencia verdadera de medias.
    - n: Tamaño de la muestra por grupo.
    - sd: Desviación estándar de ambos grupos.
    - alpha: Nivel de significancia.

    Retorna:
    - IC_lower: Límite inferior del IC.
    - IC_upper: Límite superior del IC.
    - diff: Diferencia observada de medias.
    """
    # Generar datos para el grupo control y experimental
    control = np.random.normal(loc=0, scale=sd, size=n)
    experimental = np.random.normal(loc=true_diff, scale=sd, size=n)

    # Calcular medias y desviaciones estándar
    mean_ctrl = np.mean(control)
    mean_exp = np.mean(experimental)
    sd_ctrl = np.std(control, ddof=1)
    sd_exp = np.std(experimental, ddof=1)

    # SD combinada (Pooled SD)
    pooled_sd = np.sqrt(((n - 1)*sd_ctrl**2 + (n - 1)*sd_exp**2) / (2*n - 2))

    # Diferencia de medias
    diff = mean_exp - mean_ctrl

    # Error estándar de la diferencia
    se_diff = pooled_sd * np.sqrt(2/n)

    # Grados de libertad
    df = 2*n - 2

    # Valor crítico t para el IC del 95%
    t_crit = stats.t.ppf(1 - alpha/2, df)

    # Intervalo de confianza
    IC_lower = diff - t_crit * se_diff
    IC_upper = diff + t_crit * se_diff

    return IC_lower, IC_upper, diff

# Función para generar múltiples ICs
def generar_ic_multiple(true_diff, num_sim=30, n=30, sd=1, alpha=0.05):
    """
    Genera múltiples intervalos de confianza para un verdadero valor de diferencia de medias.

    Parámetros:
    - true_diff: Diferencia verdadera de medias.
    - num_sim: Número de simulaciones.
    - n: Tamaño de la muestra por grupo.
    - sd: Desviación estándar de ambos grupos.
    - alpha: Nivel de significancia.

    Retorna:
    - Listas de límites inferiores y superiores de los ICs.
    """
    IC_lower_list = []
    IC_upper_list = []
    diffs = []
    for _ in range(num_sim):
        IC_lower, IC_upper, diff = simular_ic(true_diff, n, sd, alpha)
        IC_lower_list.append(IC_lower)
        IC_upper_list.append(IC_upper)
        diffs.append(diff)
    return IC_lower_list, IC_upper_list, diffs

# Configuración de los escenarios
escenarios = {
    'Escenario 1: Parámetro = 0.5': 0.5,
    'Escenario 2: Parámetro = 0': 0
}

# Generar los ICs para cada escenario
resultados = {}
np.random.seed(42)  # Para reproducibilidad

for nombre, valor in escenarios.items():
    IC_lower, IC_upper, diffs = generar_ic_multiple(true_diff=valor, num_sim=30, n=30, sd=1, alpha=0.05)
    resultados[nombre] = {
        'IC_lower': IC_lower,
        'IC_upper': IC_upper,
        'diffs': diffs
    }

# Función para plotear los ICs con marcas A y B, cambio de color y puntos de diferencia observada
def plot_ic_con_marcas_colores_puntos(resultados, escenarios):
    """
    Plotea los intervalos de confianza para cada escenario, añade marcas A o B
    cuando el IC no contiene el valor verdadero, cambia el color de dichos intervalos,
    y agrega la diferencia observada de medias como un punto.

    Parámetros:
    - resultados: Diccionario con los ICs y diferencias.
    - escenarios: Diccionario con los nombres y valores verdaderos.
    """
    num_escenarios = len(escenarios)
    fig, axes = plt.subplots(1, num_escenarios, figsize=(20, 20), sharey=False)

    if num_escenarios == 1:
        axes = [axes]

    for ax, (nombre, valor) in zip(axes, escenarios.items()):
        IC_lower = resultados[nombre]['IC_lower']
        IC_upper = resultados[nombre]['IC_upper']
        diffs = resultados[nombre]['diffs']

        # Crear una lista de posiciones
        posiciones = np.arange(1, len(IC_lower)+1)

        # Plotear líneas para cada IC
        for pos, lower, upper, diff in zip(posiciones, IC_lower, IC_upper, diffs):
            # Verificar si el IC contiene el valor verdadero
            contiene = lower <= valor <= upper
            if contiene:
                color = 'blue'
                marca = ''
            else:
                color = 'red'
                marca = 'A' if 'Escenario 1' in nombre else 'B'

            # Dibujar el intervalo
            ax.plot([lower, upper], [pos, pos], color=color, linewidth=2)

            # Dibujar el punto de la diferencia observada
            ax.plot(diff, pos, marker='o', color='black', markersize=6)

            # Añadir la marca si no contiene el valor verdadero
            if not contiene:
                # Determinar la posición de la marca: al final del intervalo que está más lejos del valor verdadero
                if valor < lower:
                    x_marca = lower
                else:
                    x_marca = upper
                # Añadir la marca con un desplazamiento vertical para evitar solapamientos
                ax.text(x_marca, pos - 0.2, marca, color='darkred', fontsize=14, fontweight='bold',
                        ha='center', va='bottom')

        # Línea del valor verdadero
        ax.axvline(x=valor, color='green', linestyle='--', linewidth=3, label='Valor Verdadero')

        # Configuración del eje
        ax.set_title(nombre, fontsize=22)
        ax.set_xlabel('Diferencia de Medias', fontsize=18)
        ax.set_ylabel('Simulación', fontsize=18)
        ax.set_ylim(0, len(IC_lower)+2)
        ax.set_xlim(min(IC_lower) - 1, max(IC_upper) + 1)
        ax.invert_yaxis()
        ax.legend(fontsize=16)

        # Mejorar la legibilidad de las marcas
        ax.tick_params(axis='both', which='major', labelsize=16)

    plt.suptitle('El Baile de los Intervalos de Confianza para la Diferencia de Medias (30 Simulaciones)', fontsize=28, y=0.95)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Plotear los ICs con marcas A y B, cambio de color y puntos de diferencia observada
plot_ic_con_marcas_colores_puntos(resultados, escenarios)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Configuración de estilo utilizando Seaborn
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 18})  # Aumentar tamaño de fuente para mejor legibilidad

# Parámetros de la población
mu = 50      # Media poblacional
sigma = 10   # Desviación estándar poblacional
alpha = 0.05 # Nivel de significancia para el 95% de confianza

# Definir los tres estudios con diferentes tamaños de muestra
estudios = {
    'Estudio 1': {'n': 30},
    'Estudio 2': {'n': 100},
    'Estudio 3': {'n': 500}
}

# Diccionario para almacenar los resultados
resultados = {}

# Simular cada estudio
np.random.seed(42)  # Para reproducibilidad

for nombre, params in estudios.items():
    n = params['n']
    # Simular datos
    muestra = np.random.normal(loc=mu, scale=sigma, size=n)
    # Calcular la media muestral
    media_muestral = np.mean(muestra)
    # Calcular el error estándar
    se = sigma / np.sqrt(n)
    # Calcular el intervalo de confianza
    t_crit = stats.t.ppf(1 - alpha/2, df=n-1)  # Valor crítico t
    IC_lower = media_muestral - t_crit * se
    IC_upper = media_muestral + t_crit * se
    # Almacenar los resultados
    resultados[nombre] = {
        'n': n,
        'media_muestral': media_muestral,
        'IC_lower': IC_lower,
        'IC_upper': IC_upper
    }

# Preparar los datos para la visualización
labels = list(resultados.keys())
n_values = [resultados[label]['n'] for label in labels]
IC_lowers = [resultados[label]['IC_lower'] for label in labels]
IC_uppers = [resultados[label]['IC_upper'] for label in labels]
medias = [resultados[label]['media_muestral'] for label in labels]

# Crear el gráfico
plt.figure(figsize=(12, 8))

# Usar barras horizontales para representar los IC
for i, label in enumerate(labels):
    plt.hlines(y=i, xmin=IC_lowers[i], xmax=IC_uppers[i], color='skyblue', linewidth=4)
    plt.plot(medias[i], i, 'o', color='navy', markersize=10)  # Punto de la media muestral
    plt.text(IC_lowers[i] - 1, i, f"n={n_values[i]}", color='darkblue', va='center', fontsize=12)

# Configuración del eje y
plt.yticks(range(len(labels)), labels)

# Etiquetas y título
plt.xlabel('Diferencia de Medias')
plt.title('Intervalos de Confianza de la Media según el Tamaño de la Muestra')

# Ajustar los límites del eje x para mejor visualización
plt.xlim(min(IC_lowers) - 5, max(IC_uppers) + 5)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuración de estilo utilizando Seaborn
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 12})  # Aumentar tamaño de fuente para mejor legibilidad

# Semilla para reproducibilidad
np.random.seed(42)

# 1. Distribución Normal Perfecta
normal_perfecta = np.random.normal(loc=50, scale=10, size=1000)

# 2. Distribución Normal Sesgada a la Derecha
# Generamos una distribución normal y añadimos una cola a la derecha
normal_sesgada = np.concatenate([
    np.random.normal(loc=50, scale=10, size=950),
    np.random.normal(loc=80, scale=5, size=50)
])

# 3. Distribución Normal con Outlier
# Generamos una distribución normal y añadimos outliers
# Ajustamos para que la longitud total sea 1000 (995 normales + 5 outliers)
normal_outlier = np.random.normal(loc=50, scale=10, size=995)
outliers = np.random.normal(loc=150, scale=5, size=5)  # Outliers
normal_outlier = np.concatenate([normal_outlier, outliers])

# 4. Distribución Bimodal
# Generamos dos distribuciones normales diferentes y las combinamos
bimodal = np.concatenate([
    np.random.normal(loc=40, scale=5, size=500),
    np.random.normal(loc=60, scale=5, size=500)
])

# Crear un DataFrame con todas las distribuciones
df = pd.DataFrame({
    'Normal Perfecta': normal_perfecta,
    'Normal Sesgada a la Derecha': normal_sesgada,
    'Normal con Outlier': normal_outlier,
    'Bimodal': bimodal
})

def calcular_estadisticas(data):
    """
    Calcula la media, mediana y desviación estándar de una serie de datos.

    Parámetros:
    - data: Serie de pandas o array de numpy.

    Retorna:
    - dict: Diccionario con las estadísticas calculadas.
    """
    stats_dict = {
        'Media': np.mean(data),
        'Mediana': np.median(data),
        'Desviación Estándar': np.std(data, ddof=1)
    }
    return stats_dict

def crear_graficos_rejilla(df, distribuciones, tipos_graficos):
    """
    Crea una rejilla de gráficos para las distribuciones dadas.

    Parámetros:
    - df: DataFrame de pandas que contiene las distribuciones.
    - distribuciones: Lista de nombres de las distribuciones.
    - tipos_graficos: Lista de tipos de gráficos a crear.
    """
    num_distribuciones = len(distribuciones)
    num_graficos = len(tipos_graficos)

    # Crear una figura con subplots organizados por filas y columnas
    fig, axes = plt.subplots(num_graficos, num_distribuciones, figsize=(5*num_distribuciones, 4*num_graficos))
    fig.suptitle('Visualización de Diferentes Distribuciones de Datos', fontsize=20, y=0.92)

    for col, distribucion in enumerate(distribuciones):
        data = df[distribucion]
        stats = calcular_estadisticas(data)

        for row, tipo in enumerate(tipos_graficos):
            ax = axes[row, col] if num_graficos > 1 else axes[col]

            if tipo == 'Densidad':
                sns.kdeplot(data, ax=ax, shade=True, color='purple')
                ax.axvline(stats['Media'], color='red', linestyle='--', linewidth=2, label=f'Media: {stats["Media"]:.2f}')
                ax.axvline(stats['Mediana'], color='green', linestyle='-', linewidth=2, label=f'Mediana: {stats["Mediana"]:.2f}')
                # Desviación estándar como sombreado
                ax.axvspan(stats['Media'] - stats['Desviación Estándar'],
                           stats['Media'] + stats['Desviación Estándar'],
                           color='grey', alpha=0.2, label='1 SD')
                ax.set_title(f'{distribucion} - {tipo}')
                ax.legend()

            elif tipo == 'Histograma':
                sns.histplot(data, bins=30, ax=ax, kde=True, color='orange')
                ax.axvline(stats['Media'], color='red', linestyle='--', linewidth=2, label=f'Media: {stats["Media"]:.2f}')
                ax.axvline(stats['Mediana'], color='green', linestyle='-', linewidth=2, label=f'Mediana: {stats["Mediana"]:.2f}')
                # Desviación estándar como sombreado
                ax.axvspan(stats['Media'] - stats['Desviación Estándar'],
                           stats['Media'] + stats['Desviación Estándar'],
                           color='grey', alpha=0.2, label='1 SD')
                ax.set_title(f'{distribucion} - {tipo}')
                ax.legend()

            elif tipo == 'Boxplot':
                sns.boxplot(x=data, ax=ax, color='lightgreen')
                ax.axvline(stats['Media'], color='red', linestyle='--', linewidth=2, label=f'Media: {stats["Media"]:.2f}')
                ax.axvline(stats['Mediana'], color='green', linestyle='-', linewidth=2, label=f'Mediana: {stats["Mediana"]:.2f}')
                ax.set_title(f'{distribucion} - {tipo}')
                ax.legend()

    plt.tight_layout(rect=[0, 0, 1, 0.90])
    plt.show()

# Definir las distribuciones y tipos de gráficos
distribuciones = ['Normal Perfecta', 'Normal Sesgada a la Derecha', 'Normal con Outlier', 'Bimodal']
tipos_graficos = ['Densidad', 'Histograma', 'Boxplot']

# Crear los gráficos
crear_graficos_rejilla(df, distribuciones, tipos_graficos)
```
